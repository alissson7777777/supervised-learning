# Проект: Обучение с учителем – Качество модели

    Проект направлен на разработку модели, которая позволит персонализировать предложения для постоянных клиентов интернет-магазина «В один клик» с целью повышения их покупательской активности. 
---

## Описание проекта
 
Интернет-магазин «В один клик» столкнулся с проблемой снижения активности постоянных покупателей. Новые клиенты уже не привлекают магазин, поэтому важным становится удержание старых. Решением является создание персонализированных предложений для клиентов, чья активность снижалась.

**Задача:**  
- **Целевой признак:** Покупательская активность, которая принимает два значения – «снизилась» и «прежний уровень».
- **Цели модели:** Предсказать вероятность того, что покупатель в следующие 3 месяца будет покупать меньше (то есть активность снизится).

**Дополнительные данные:**  
Помимо исходных данных о поведении покупателей (коммуникации, продуктовое поведение, покупательское поведение, поведение на сайте), в исследование включены данные о прибыльности клиента за последние три месяца, предоставленные финансовым департаментом.

---

## Цель проекта

Цель проекта — разработать и внедрить модель машинного обучения, которая:
- Прогнозирует вероятность снижения покупательской активности постоянного клиента.
- Используя прогнозы и данные о прибыльности, позволяет сегментировать клиентов для разработки персонализированных предложений с целью увеличения их активности.

````python
market_file_dtypes = {
    'id':'int64',
    'маркет_актив_6_мес': 'float64',
    'акционные_покупки_6_мес,%': 'float64',
    'зарегестрирован,дни': 'int64',
    'ср_просмотр_категорий_за_визит': 'int64',
    'неоплаченное_3_мес,шт': 'int64',
    'ошибка_сервиса': 'int64',
    'стр_за_визит_3_мес':'int64'
}

market_money_dtypes = {
    'id':'int64',
    'выручка': 'float64'
}

market_time_dtypes = {
    'id':'int64',
    'время_на_сайте,мин':'float64'
}


money_dtypes = {
    'id':'int64',
    'прибыль': 'float64'
}

# создадим функцию для перевода столбцов в нужные типы
def convert_dtypes(df, dtype_dict):
    for col, desired_type in dtype_dict.items():
        df[col] = pd.to_numeric(df[col], errors='coerce')
        
        if 'int' in desired_type:
            df[col] = df[col].astype('Int64')
        else:
            df[col] = df[col].astype(desired_type)

convert_dtypes(market_file, market_file_dtypes)
convert_dtypes(market_money, market_money_dtypes)
convert_dtypes(market_time, market_time_dtypes)
convert_dtypes(money, money_dtypes)
````

**Проверим наличие явных дубликатов** 

Создадим функцию для отображения и удаления явных дубликатов в каждом датафрейме.

````python
def dup_entities_multiple(*dfs, subset=None, keep="first"):
    cleaned = []
    for i, df in enumerate(dfs, start=1):
        num_duplicates = df.duplicated(subset=subset, keep=keep).sum()
        print(f"[DataFrame {i}] Количество дубликатов: {num_duplicates}")
        if num_duplicates > 0:
            df = df.drop_duplicates(subset=subset, keep=keep)
            print(f"[DataFrame {i}] Дубликаты удалены.")
        else:
            print(f"[DataFrame {i}] Дубликаты отсутствуют.")
        cleaned.append(df)
    return cleaned


market_file, market_time, market_money, money = dup_entities_multiple(
    market_file, 
    market_time, 
    market_money, 
    money
)
````

Создадим функции для подчета статистических показателей. 

````python
#для определения границ ящика с учами воспользуемся формулой для расчета границ выбросов с использованием
#межквартильного размаха (IQR). Создадим функцию, которую будем в дальнейшем не раз вызывать.
def borders(data,column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    # границы для выбросов
    lower_bound = round(Q1 - 1.5 * IQR,4)
    upper_bound = round(Q3 + 1.5 * IQR,4)
    print(f"Размах \"усов\": \n1)нижняя граница: {lower_bound}\n2)верхняя граница: {upper_bound}")
````

````python
#создадим функцию для отображения статистических показателей, гистограммы и ящика с усами
def qunatity_analisys(df,col,bins=30):
    display(df[col].describe().round(3))
    print(f'\n')
    
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))  
    df[col].hist(bins = bins, ax = axes[0]);
    df.boxplot(column= col, ax = axes[1]);

    axes[0].set_title('Гистограмма распределениея', fontsize=12)
    axes[0].set_xlabel(col, fontsize=10)
    axes[0].set_ylabel('Количество наблюдений', fontsize=10)

    axes[1].set_title('«График „ящик с усами“»', fontsize=12)
#     axes[1].set_ylabel(col, fontsize=10)

    plt.show()
````

````python
#создадим функцию для сравнения количественных признаков
def col_comparison(df_1, df_2, col_1, col_2,bins=30):
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))                        
    axes[0].hist(df_1[col_1], bins=bins, alpha=0.6, label=col_1, color='blue', edgecolor='black')
    axes[0].hist(df_2[col_2], bins=bins, alpha=0.6, label=col_2, color='orange', edgecolor='black')
    axes[0].set_title('Гистограмма распределения', fontsize=16)
#     axes[0].set_xlabel(col, fontsize=14)
    axes[0].set_ylabel('Количество наблюдений', fontsize=14)
    axes[0].legend(loc='upper right', fontsize=12)
    axes[0].grid(axis='y', alpha=0.75)
                             
    box_data = [df_1[col_1], df_2[col_2]]
    axes[1].boxplot(box_data, labels=[col_1, col_2])
    axes[1].set_title('Ящик с усами', fontsize=16)
#     axes[1].set_ylabel(col, fontsize=14)
    axes[1].grid(axis='y', alpha=0.75)

    plt.tight_layout()  
    plt.show()
    
    qunatity_analisys(df_1,col_1,bins=bins)
    qunatity_analisys(df_2,col_2,bins=bins)
````

````python
#создадим функцию для хар-ки категориальных признаков
def cat_characteristic(df, col, title_name=None,figsize=(6, 6)):
    if col not in df.columns:
        raise ValueError(f"Столбец '{col}' отсутствует в датафрейме.")
    
    counts = df[col].value_counts()
    
    def func(pct, allvals):
        absolute = round(pct / 100.*sum(allvals))
        return f"{pct:.1f}%\n({absolute})"
    
    counts.plot(
        kind='pie',
        autopct=lambda pct: func(pct, counts),
        startangle=90,
        figsize=figsize,
        colors=plt.cm.Paired.colors
    )
    plt.title(title_name, fontsize=14)
    plt.ylabel('') 
    plt.tight_layout()
    plt.show()
````

````python
# создаём пайплайн для подготовки признаков из списка ord_columns: заполнение пропусков и Ordinal-кодирование
# SimpleImputer + OrdinalEncoder
ord_pipe = Pipeline(
    [
        (
            'simpleImputer_before_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',
            OrdinalEncoder(
                categories=[['Слабая', 'Средняя', 'Высокая'],
                            ['стандарт','премиум']
                ], 
                handle_unknown='use_encoded_value',
                unknown_value=np.nan
            )
        ),
        (
            'simpleImputer_after_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)
````

````python
num_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),  
    ('scaler', StandardScaler()), 
    ('min_max', MinMaxScaler())  
])
````

````python
# создаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', num_pipe, num_columns)
    ], 
    remainder='passthrough'
)
````

````python
pipe_final = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
    ]
)
````

````python
def save_model(model, model_name):
    filename = f"best_{model_name}_model.pkl"
    joblib.dump(model, filename)
    print(f"Модель сохранена в {filename}")
````

Создадим функцию, которая будет сразу сегментировать покупательскую активность и выбранный признак. 

````python
def segment_and_visualize(segment_df,bins_activity, labels_activity,column_name, title, ylabel):
    
    if 'Сегмент активности' not in segment_df.columns:
        segment_df['Сегмент активности'] = pd.cut(
            segment_df['вероятность_снижения_активности'], bins=bins_activity, labels=labels_activity
        ).astype(str)
    
    median_value = segment_df[column_name].median()
    
    segment_df[f'Сегмент {column_name}'] = np.where(segment_df[column_name] > median_value, 'высокая', 'низкая')
    

    segment_df['Финальный сегмент'] = segment_df['Сегмент активности'].astype(str) + ' | ' + segment_df[f'Сегмент {column_name}']
    

    plt.figure(figsize=(12, 6))
    colors = {
        'низкая | низкая': 'blue', 'низкая | высокая': 'green',
        'средняя | низкая': 'orange', 'средняя | высокая': 'yellow',
        'высокая | низкая': 'purple', 'высокая | высокая': 'red'
    }
    
    for label, color in colors.items():
        subset = segment_df[segment_df['Финальный сегмент'] == label]
        plt.scatter(subset['вероятность_снижения_активности'], subset[column_name], label=label, color=color, alpha=0.6)
    
    plt.xlabel('Вероятность снижения активности')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(title="Сегмент")
    plt.grid(True)
    plt.show()
````
